```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/car_crash_data.csv")
df.head()
```

# Alcohol vs Total Crash Rate Analysis

## Research Question
Does alcohol involvement have a measurable relationship with total crash percentage across U.S. states?

This exploratory data analysis investigates whether higher alcohol-related crash percentages are associated with higher overall crash rates.

## Correlation Analysis

To determine the strength of the relationship between alcohol involvement and total crash percentage, we calculate the Pearson correlation coefficient.

```python
correlation = df["alcohol"].corr(df["total"])
print("Correlation between Alcohol and Total Crash Percentage:", correlation)
```

## Scatterplot Visualization

A scatterplot helps visualize the relationship between alcohol-related crashes and total crash percentage.

```python
plt.figure()
sns.scatterplot(data=df, x="alcohol", y="total")
plt.title("Alcohol vs Total Crash Percentage by State")
plt.xlabel("Alcohol-Related Crash Percentage")
plt.ylabel("Total Crash Percentage")
plt.show()
```

```python
plt.figure()
sns.regplot(data=df, x="alcohol", y="total")
plt.title("Alcohol vs Total Crash Percentage with Trend Line")
plt.xlabel("Alcohol-Related Crash Percentage")
plt.ylabel("Total Crash Percentage")
plt.show()
```

## States with Highest Alcohol-Related Crash Percentages

Identifying the states with the highest alcohol-related crash percentages.

```python
df.sort_values("alcohol", ascending=False).head()
```

## Conclusion

The correlation analysis shows a positive relationship between alcohol involvement and total crash percentage. 
The scatterplot and regression line indicate that as alcohol-related crash percentages increase, total crash percentages also tend to increase.

Although alcohol is not the only contributing factor, it appears to be a meaningful predictor of crash rates across states.
